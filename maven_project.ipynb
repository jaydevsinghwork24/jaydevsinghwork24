```python
for i in dfcalc.index:
    dfcalc.loc[i,'total_cost']=dfcalc.loc[i,'total_transaction']*dfcalc.loc[i,'product_cost']
    dfcalc.loc[i,'expected_revenue']=dfcalc.loc[i,'total_transaction']*dfcalc.loc[i,'product_retail_price']
    dfcalc.loc[i,'total_loss']=dfcalc.loc[i,'total_return']*dfcalc.loc[i,'product_cost']
    dfcalc.loc[i,'profit']=dfcalc.loc[i,'expected_revenue']-dfcalc.loc[i,'total_loss']
    
dfcalc
```
